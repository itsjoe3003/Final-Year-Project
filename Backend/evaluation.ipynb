```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
metrics = ('Engagement', 'Retention', 'Progression', 'Motivation')
values = {
    'Conventional Approach': (4,3,5,4),
    'Gamified Approach': (9,8,8,7)
}
```

```python
x = np.arange(len(metrics))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0
```

```python
fig, ax = plt.subplots(layout='constrained')

for approach, value in values.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, value, width, label=approach)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Values')
ax.set_title('Conventional vs Gamified Learning Approach')
ax.set_xticks(x + width, metrics)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 12)

plt.show()
```
